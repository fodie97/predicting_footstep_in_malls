```python
# 🔹 Display column names to check for errors
print("Columns in dataset:", df_final2.columns)

# 🔹 Ensure 'Jour' exists in dataset
if 'Jour' not in df_final2.columns:
    raise KeyError("❌ 'Jour' column not found in the dataset! Check the column names.")

# 🔹 Convert 'Jour' to datetime
df_final2['Jour'] = pd.to_datetime(df_final2['Jour'], errors='coerce')

# 🔹 Check for failed conversions (NaT values)
num_invalid_dates = df_final2['Jour'].isna().sum()
if num_invalid_dates > 0:
    print(f"⚠ Warning: {num_invalid_dates} rows have invalid 'Jour' values and will be removed!")

# 🔹 Drop rows where 'Jour' could not be converted
df_final2 = df_final2.dropna(subset=['Jour'])

# 🔹 Check if 'Entrées' exists
if 'Entrées' not in df_final2.columns:
    raise KeyError("❌ 'Entrées' column not found in the dataset! Check the column names.")

# 🔹 Aggregate by day while keeping relevant data
df_daily = df_final2.groupby('Jour', as_index=False).agg({
    'Entrées': 'sum',  # Sum of entries per day
    'Mall ID': 'first' if 'Mall ID' in df.columns else 'mean',
    'Année': 'first' if 'Année' in df.columns else 'mean',
    'Mois': 'first' if 'Mois' in df.columns else 'mean',
    'jour_semaine': 'first' if 'jour_semaine' in df.columns else 'mean',
    'GLA centre (hyper + galerie+ mail) GI': 'first' if 'GLA centre (hyper + galerie+ mail) GI' in df.columns else 'mean',
    'GLA boutiques (hors mail - GI)': 'first' if 'GLA boutiques (hors mail - GI)' in df.columns else 'mean',
    'Nombre places de parking (Assetbook)': 'first' if 'Nombre places de parking (Assetbook)' in df.columns else 'mean',
    'Nombre de boutiques (hors hyper)': 'first' if 'Nombre de boutiques (hors hyper)' in df.columns else 'mean',
    'ALIMENTATION': 'first' if 'ALIMENTATION' in df.columns else 'mean',
    'EQ DE LA PERSONNE': 'first' if 'EQ DE LA PERSONNE' in df.columns else 'mean',
    'EQ DE LA MAISON': 'first' if 'EQ DE LA MAISON' in df.columns else 'mean',
    'LOISIRS': 'first' if 'LOISIRS' in df.columns else 'mean',
    'RESTAURATION': 'first' if 'RESTAURATION' in df.columns else 'mean',
    'SERVICES': 'first' if 'SERVICES' in df.columns else 'mean',
    'DIVERS': 'first' if 'DIVERS' in df.columns else 'mean'
})



print("✅ Aggregation completed successfully!")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Identify numerical columns for normalization (excluding target variable 'Entrées')
num_features = [
    'GLA centre (hyper + galerie+ mail) GI',
    'GLA boutiques (hors mail - GI)',
    'Nombre places de parking (Assetbook)',
    'Nombre de boutiques (hors hyper)',
    'ALIMENTATION', 'EQ DE LA PERSONNE', 'EQ DE LA MAISON',
    'LOISIRS', 'RESTAURATION', 'SERVICES', 'DIVERS'
]

# Check which columns exist in df_daily to avoid errors
num_features = [col for col in num_features if col in df_daily.columns]

# Initialize scalers
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()

# Normalize: Use StandardScaler for normally distributed columns, MinMaxScaler for skewed ones
for feature in num_features:
    if df_daily[feature].nunique() > 1:  # Check if column has more than one unique value
        if df_daily[feature].skew() < 1:  # Normal distribution
            df_daily[feature] = scaler_standard.fit_transform(df_daily[[feature]])
        else:  # Skewed distribution
            df_daily[feature] = scaler_minmax.fit_transform(df_daily[[feature]])
    else:
        print(f"⚠ Skipping normalization for '{feature}' (only one unique value)")
```

```python
from sklearn.preprocessing import OneHotEncoder

# Identify categorical columns for encoding
categorical_features = ['jour_semaine']

# Check which categorical features exist in df_daily
categorical_features = [col for col in categorical_features if col in df_daily.columns]

# Apply OneHotEncoding with the correct parameter for newer versions of sklearn
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Use sparse_output instead of sparse

# Transform categorical features
encoded_array = encoder.fit_transform(df_daily[categorical_features])

# Create a DataFrame with encoded features
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_features))

# Merge with original dataset and drop old categorical columns
df_daily = pd.concat([df_daily.drop(columns=categorical_features), encoded_df], axis=1)
```
